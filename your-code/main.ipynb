The computation of the mean might give us a clue regarding how the statistical test may turn out; However, it certainly does not prove whether there is a significant difference between the two groups.

In the cell below, use the `ttest_ind` function in `scipy.stats` to compare the the total points for legendary and non-legendary Pokemon. Since we do not have any information about the population, assume the variances are not equal.

```python
#1. Set the hypothesis

## H0: mu is the total stats of legendary pokemons = total stats of non legendary pokemons
## H1: mu is the total stats of legendary pokemons != total stats of non legendary pokemons

## We are dealing with a two-tailed test
```

```python
#2. Identify the significance level

alpha = 0.05
```

```python
#3. Determine the sample
```

```python
legendaries = pokemon[pokemon['Legendary']==True]['Total']
legendaries.head(5)
```

```python
non_legendaries = pokemon[pokemon['Legendary']==False]['Total']
non_legendaries.head(5)
```

```python
#4. Compute the statistics
#5. Calulate the p-value

stat, p_value = st.ttest_ind(legendaries, non_legendaries, equal_var=False)

print(stat)
print(p_value) 
```

```python
#5. Make a decision

p_value < alpha
```

What do you conclude from this test? Write your conclusions below.

#### Your conclusions here:

The p_value is smaller than alpha, therefore we can reject H0 and do not reject H1, or the mu total stats of legendary pokemon will be more likely different from the non-legendary ones.

# Challenge 2 - Matched Pairs Test

In this challenge we will compare dependent samples of data describing our Pokemon. Our goal is to see whether there is a significant difference between each Pokemon's defense and attack scores. Our hypothesis is that the defense and attack scores are equal. In the cell below, import the `ttest_rel` function from `scipy.stats` and compare the two columns to see if there is a statistically significant difference between them.

```python
pokemon.head(10)
```

```python
#1. Set hypothesis (H0 and H1)
# H0: mu attack stats for each pokemon = mu defense stats for each pokemon
# H1: mu attack stats for each pokemon != mu defense stats for each pokemon

#2. Identify significance level
alpha = 0.05 
```

```python
#3. Define the sample

pokemon_atk = pokemon['Attack']
pokemon_def = pokemon['Defense']
```

```python
#4. Compute the statistics
#5. Calculate the p-vale
#6. Make a decision


stat, p_value = st.ttest_rel(pokemon_atk, pokemon_def) 

print(stat)
print(p_value) 

p_value < alpha
```

Describe the results of the test in the cell below.

#### Your conclusions here:

Our p_value is smaller than our alpha, so we can reject H0 and start believing that the mu attack statistics for each pokemon are different to the mu defense statistics for each pokemon. This means that we do not reject H1.

# Bonus Challenge - The Chi-Square Test

The Chi-Square test is used to determine whether there is a statistically significant difference in frequencies. In other words, we are testing whether there is a relationship between categorical variables or rather when the variables are independent. This test is an alternative to Fisher's exact test and is used in scenarios where sample sizes are larger. However, with a large enough sample size, both tests produce similar results. Read more about the Chi Squared test [here](https://en.wikipedia.org/wiki/Chi-squared_test).

In the cell below, create a contingency table using `pd.crosstab` comparing whether a Pokemon is legenadary or not and whether the Type 1 of a Pokemon is water or not.

```python
from scipy.stats import chisquare
```

```python
pokemon.head(10)
```

```python
pokemon_1 = pd.crosstab(pokemon['Type 1'], pokemon['Legendary'])
pokemon_1
```

```python
#1. Set hypothesis (H0 and H1)
# H0: mu statistically significant in frequencies is equal between Type 1 and Legendary, not Legendary
# H1: mu statistically significant in frequencies is different between Type 1 and Legendary, not Legendary

#2. Identify significance level
alpha = 0.05
```

Perform a chi-squared test using the `chi2_contingency` function in `scipy.stats`. You can read the documentation of the function [here](https://docs.scipy.org/doc/scipy-0.15.1/reference/generated/scipy.stats.chi2_contingency.html).

```python
#3. Define the sample --> pokemon_1

#4. Compute the statistics
#5. Compute the p-values

stat, p_value = chisquare(pokemon_1, axis=0)

#The axis of the broadcast result of f_obs and f_exp along which to apply the test. If axis is None, all values in f_obs are treated as a single data set. Default is 0.

print(stat)
print(p_value)
```

Based on a 95% confidence, should we reject the null hypothesis?

```python
#6. Make a decision

## P-value of Non Legendary and Type 1

6.55748731e-58 < alpha
```

```python
## P-value of Legendary and Type 1

8.43266971e-08 < alpha
```

#### Comments:

Both p-values are smaller than our significance level of 5% or confidence level of 95%. Hence, we reject H0 and start to believe that statistically significant in frequencies is different between Type 1 and Legendary, not Legendary.
